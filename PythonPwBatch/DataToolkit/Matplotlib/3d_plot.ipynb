```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)
```

```python
plt.figure().add_subplot(projection='3d')
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.plot(x,y,z, color='r')
plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.plot(x,y,z, color='r')
plt.xlabel('jghjg')
plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.bar(x,y,z, color='r')
plt.xlabel('jghjg')
plt.show()
```
