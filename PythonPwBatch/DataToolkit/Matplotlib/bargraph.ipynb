```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ['a','b','c','d','e']
y = [4,6,1,2,3]
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y, color='g')
plt.xlabel('alphabet')
plt.ylabel('frequency')
plt.title('BAR GRAPH')
plt.show()
```

```python
plt.bar(x,y, color='red')
plt.xlabel('alphabet')
plt.ylabel('frequency')
plt.title('BAR GRAPH')
plt.show()
```
