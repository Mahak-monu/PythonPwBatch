```python
data1 = np.random.normal(0,1,5000)
data2 = np.random.normal(0,1,5000)
```

```python
data1
```

```python
plt.hist([data1, data2])
plt.show()
```

```python
plt.hist([data1, data2], bins=20)
plt.show()
```

```python
plt.hist([data1, data2], color=['r','g'], bins=20)
plt.show()
```

```python
plt.hist([data1, data2], color=['r','g'],label =['dist1', 'dist2'], bins=20)
plt.legend()
plt.show()
```

```python
plt.hist([data1, data2], color=['r','g'],label =['dist1', 'dist2'], bins=20)
plt.legend()
plt.title('stacked histogram')
plt.xlabel('values')
plt.ylabel('frequency')
plt.show()
```
