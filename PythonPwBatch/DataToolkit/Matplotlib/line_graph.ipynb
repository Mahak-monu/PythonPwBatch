```python
x=[1,2,3,4,5]
y1=[5,1,6,8,2]
y2 = [7,8,3,5,1]
```

```python
plt.plot(x,y1, color='r', marker='o', markersize=10, linestyle='--', linewidth=3, label='line y1')
plt.legend()
```

```python
plt.plot(x,y1, color='r', marker='o', markersize=10, linestyle='--', linewidth=3, label='line y1')
plt.plot(x,y2, color='b', marker='o', markersize=10, linestyle='--', linewidth=3, label='line y2')
plt.xlabel('x data axis')
plt.ylabel('y label axis')
plt.title('multiple  lines in one  plot')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
plt.show()
```

```python
plt.figure()
plt.subplot(2,2, 1) # 2 rows 2 columns
plt.plot(x, np.sin(x))
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.subplot(2,2, 3)
plt.plot(x, np.tan(x))
plt.subplot(2,2,4)
plt.plot(x,np.cos(x+1))
plt.show()
```

```python
plt.figure()
plt.subplot(1,4, 1)       # 1 row and 4 columns
plt.plot(x, np.sin(x))
plt.subplot(1,4,2)
plt.plot(x,np.cos(x))
plt.subplot(1,4, 3)
plt.plot(x, np.tan(x))
plt.subplot(1,4,4)
plt.plot(x,np.cos(x+1))
plt.show()
```

```python
plt.figure()
plt.subplot(4,1, 1)  # 4 rows and 1 column
plt.plot(x, np.sin(x))
plt.subplot(4,1,2)
plt.plot(x,np.cos(x))
plt.subplot(4,1, 3)
plt.plot(x, np.tan(x))
plt.subplot(4,1,4)
plt.plot(x,np.cos(x+1))
plt.show()
```
