```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
langs = ['C','Python','Java']
students = [20,100,40]
```

```python
plt.pie(students, labels=langs, autopct='%1.1f%%')
plt.title('pie chart of prog lang')
plt.show()
```

```python
explode=[0.0,0.1,0.2]
```

```python
plt.pie(students, labels=langs, autopct='%1.1f%%', explode=explode)
plt.title('pie chart of prog lang')
plt.show()
```

```python
plt.pie(students, labels=langs, autopct='%1.1f%%', explode=explode, shadow=True)
plt.title('pie chart of prog lang')
plt.show()
```

```python
plt.pie(students, labels=langs, autopct='%1.1f%%',  shadow=True)
plt.title('pie chart of prog lang')
plt.show()
```
