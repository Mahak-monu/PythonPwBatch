```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.Fare
```

```python
df['new_fare'] = df['Fare'].apply(lambda x: x*90)
```

```python
df
```

```python
df['name_len'] = df['Name'].apply(len)
```

```python
df
```

```python
def convert_fare(x):
    return x*90
```

```python
df['fare1'] = df.Fare.apply(convert_fare)
```

```python
df
```

```python
def check_fare(x):
    if x<=10:
        return "cheap"
    elif x>10 and x<=30:
        return "medium"
    else:
        return "high"
```

```python
df['fare_info'] = df.Fare.apply(check_fare)
```
