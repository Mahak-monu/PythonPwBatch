# ORDINAL ENCODING
we specify labels and not the system by itself, 
there is order in the data

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df2 = pd.DataFrame({
    'qualification':['HS','GR','PG','Phd','HS','HS','Phd','PG']
})
```

```python
df2
```

```python
ordinal_encoder = OrdinalEncoder(categories=[['HS','GR','PG','Phd']])
```

```python
ordinal_encoder
```

```python
array_ordinal = ordinal_encoder.fit_transform(df2[['qualification']])
```

```python
array_ordinal
```

```python
df_ordinal = pd.DataFrame(array_ordinal, columns=['encoded_qualification'])
```

```python
df_ordinal
```

```python
df_ordinal_encoded = pd.concat([df2,df_ordinal], axis=1)
```

```python
df_ordinal_encoded
```

# TARGET GUIDED ORDINAL ENCODING

make similar category groups and then replace by mean of that group

```python
df3 = pd.DataFrame({'time':['lunch','breakfast','dinner','lunch','breakfast','dinner','lunch','breakfast','dinner','lunch','breakfast','dinner'],
                    'total_bill' : [120,130,100,100,110,200,140,70,190,200,210,90]})
```

```python
df3
```

```python
df3.groupby('time')['total_bill'].mean()
```

```python
df3.groupby('time').mean()['total_bill']
```

```python
mean_price = df3.groupby('time').mean()['total_bill'].to_dict()
```

```python
mean_price
```

```python
df3['time']
```

```python
df3['time_encoded'] = df3['time'].map(mean_price)
```
