# STANDARDIZATION

```python
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
mean = df['total_bill'].mean()
```

```python
std = df['total_bill'].std()
```

```python
mean
```

```python
std
```

```python
list(df['total_bill'])
```

```python
std_data = []
for i in list(df['total_bill']):
    std_data.append((i-mean)/std)
    
```

```python
std_data
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(std_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```

```python
scaler.fit_transform(df[['total_bill']])
```

```python
arr = scaler.fit_transform(df[['total_bill', 'tip']])
```

```python
arr
```

```python
df1 = pd.DataFrame(arr, columns=['bill','tip'])
```

```python
sns.displot(df['total_bill'], kde=True)
```

# NORMALIZATION

```python
# manually
```

```python
df
```

```python
min = df['total_bill'].min()
```

```python
min
```

```python
max = df['total_bill'].max()
```

```python
max
```

```python
df.describe()
```

```python
norm_data = []
for i in list(df['total_bill']):
    norm_data.append((i-min)/(max-min))
```

```python
norm_data
```

```python
sns.displot(df['total_bill'],kde=True)
```

```python
sns.displot(norm_data, kde=True)
```
